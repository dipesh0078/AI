#Data Cleaning & Pre-Processing

```python
df.isnull().sum()
```

```python
data=df[['x','y','SEVERITYCODE','WEATHER','ROADCOND','LIGHTCOND','SPEEDING','UNDERINFL','PERSONCOUNT','PEDCOUNT','PEDCYLCOUNT','VEHCOUNT','INJURIES','SERIOUSINJURIES','SEVERITYDESC','FATALITIES']]
```

```python
data.isnull().sum()
```

We can see speeding,weather,roadcond etc  column has
still lots of null values

```python
data['SPEEDING'].value_counts()
```

Here the y in speeding represent those who are at greater speed then the threshold and all null values refers to not speeding cases. SO converting all y to 1 and all NaN to 0.

```python
data['SPEEDING'].replace(np.nan,0,inplace=True)
data['SPEEDING'].replace('Y',1,inplace=True)
data['SPEEDING'].value_counts().to_frame('Counts')
```

```python
data["WEATHER"]
```

The dataset still have some columns like weather, roadcond etc which contains unlnown values and they are a kind of outliers so all are converted to nan

```python
data.replace(to_replace={'Unknown':np.nan,
                         'Other':np.nan},inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
#droppping all missing data from the dataset
data.isnull().sum()
```

```python
data['UNDERINFL'].value_counts()
```

```python
#changing these variable to numerica form
data['UNDERINFL'].replace(to_replace={'Y':1, 'N':0, '1':1, '0':0}, inplace=True)
```

```python
data['UNDERINFL'].value_counts()
```

```python
data.info()
```

```python
data['SERIOUSINJURIES'].value_counts()
```
