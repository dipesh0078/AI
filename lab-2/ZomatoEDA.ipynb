```python
df_country=pd.read_excel('/content/Country-Code.xlsx')
df_country.info()
```

##Combining Two data frames using pd.merge

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
#pie chart-Top 3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text'])
ratings=ratings.size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

##Observation
1. When RAting is between 4.5 to 4.9 ---> Excellent
2. When RAting are between 4.0 to 4.4 --->very good
3. When RAting is between 3.5 to 3.9---> Good
4. When rating is betwwen 3.0 to 3.4---> average
5. WHen the rating is between 2.5 to 2.9 ---> average
5. WHen the rating is between 2.0 to 2.4 ---> poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Count',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

#Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
##COunt plot
sns.countplot(x='Rating color',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```
