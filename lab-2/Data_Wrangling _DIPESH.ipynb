#7. Conduct hypothesis testing on target features using chi-square tests and test for normality using Shapiro test or Kolmogorov-Smirnov test.

Hypothesis Testing (Roadcondition and Fatalities)

```python
import scipy.stats as stats
```

```python
dataset_table=pd.crosstab(df['ROADCOND'],df['FATALITIES'])
dataset_table.values
```

```python
dataset_table
```

```python
Observed_Values = dataset_table.values
print("Observed Values :-\n",Observed_Values)
```

```python
val=stats.chi2_contingency(dataset_table)
val
```

```python
Expected_Values=val[3]
```

```python
no_of_rows=len(dataset_table.iloc[0:8,0])
no_of_columns=len(dataset_table.iloc[0,0:6])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
```

```python
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('p-value:',p_value)
```

```python
if p_value<=alpha:
    print("Reject H0,There is a relationship between   variables")
else:
    print("Retain H0,There is no relationship between  variables")
```

 # Test for normality using Shapiro test

```python
test_feature=df.INJURIES
```

```python
test_feature.dtype
```

```python
#Informal method using distplot
sns.distplot(df['INJURIES'], kde=False)
```

```python
#Null:Data is normally distributed
#ALT: Data is not normally distributed
#shapiro_wilk Test

from numpy.random import seed
from scipy.stats import shapiro
seed(1)
stat,p=shapiro(test_feature)
print('Statistics=%.3f, p=%.3f' % (stat,p))
#interpret
alpha=0.05
if p>alpha:
  print('Sample looks Gaussian(fail to reject H0)')
else:
  print('Sample does not look Gaussian(reject H0)')
```

# 8. Conduct correlation testing, using Pearson correlation test for linear relationships between features and determining the appropriate test for non-linear relationships (e.g., through visual inspection with scatterplots).

```python
#Performing Pearson Correlation:
numeric_variables = ['PERSONCOUNT', 'PEDCOUNT', 'PEDCYLCOUNT', 'VEHCOUNT', 'INJURIES', 'SERIOUSINJURIES', 'FATALITIES']
```

```python
numeric_df=df[numeric_variables]
```

```python
correlation_matrix = numeric_df.corr(method='pearson')
correlation_matrix
```

**1.PERSONCOUNT and VEHCOUNT:**
Pearson correlation coefficient: 0.566163
Interpretation: There is a moderate positive linear relationship (0.57) between the number of people involved in a collision (PERSONCOUNT) and the number of vehicles involved (VEHCOUNT). This indicates that collisions involving more people tend to involve more vehicles.<br>
**2.PERSONCOUNT and INJURIES:**
Pearson correlation coefficient: 0.323373
Interpretation: There is a moderate positive linear relationship (0.32) between the number of people involved in a collision (PERSONCOUNT) and the number of injuries (INJURIES). This suggests that collisions involving more people tend to result in more injuries.<br>
**3.INJURIES and SERIOUSINJURIES:**
Pearson correlation coefficient: 0.278052
Interpretation: There is a moderate positive linear relationship (0.28) between the total number of injuries (INJURIES) and the number of serious injuries (SERIOUSINJURIES). This implies that collisions resulting in more injuries also tend to result in more serious injuries.<br>
**4.PEDCOUNT and PEDCYLCOUNT:**
Pearson correlation coefficient: -0.017815
Interpretation: There is a very weak negative linear relationship (-0.02) between the number of pedestrians involved (PEDCOUNT) and the number of cyclists involved (PEDCYLCOUNT). This suggests that collisions involving more pedestrians do not necessarily involve more cyclists and vice versa.<br>
**5.SERIOUSINJURIES and FATALITIES:**
Pearson correlation coefficient: 0.150525
Interpretation: There is a weak positive linear relationship (0.15) between the number of serious injuries (SERIOUSINJURIES) and the number of fatalities (FATALITIES). This indicates that collisions resulting in more serious injuries also tend to result in more fatalities, although the relationship is weaker compared to others.

```python
for i in range(len(numeric_variables)):
    for j in range(i+1, len(numeric_variables)):
        plt.figure(figsize=(8, 6))
        plt.scatter(df[numeric_variables[i]], df[numeric_variables[j]], alpha=0.5)
        plt.title(f'Scatterplot of {numeric_variables[i]} vs {numeric_variables[j]}')
        plt.xlabel(numeric_variables[i])
        plt.ylabel(numeric_variables[j])
        plt.grid(True)
        plt.show()
```

While inscpecting the above scatterplots we can see that there is a cluster like formation in PERSONCOUNT VS INJURIES ans SERIOUSINJURIES,it indicates that a non-linear correlation may exist between the variables. In this case, we can consider using non-parametric correlation tests such as Spearman correlation or Kendall's tau to quantify the strength and direction of the relationship.
